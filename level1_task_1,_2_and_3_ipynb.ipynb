```python
import os

# Rename to something simpler
os.rename("Dataset  (1) (3).csv", "zomato.csv")

# Now use clean name
df = pd.read_csv("zomato.csv")
df.head()
```

```python
print("ðŸ”¹ Dataset Shape")
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])

print("\nðŸ”¹ Missing Values")
print(df.isnull().sum())
```

```python
# Drop rows with missing 'Cuisines'
df = df.dropna(subset=['Cuisines'])

print("\nðŸ”¹ Data Types")
print(df.dtypes)
```

```python
# Convert 'Votes' column to numeric if needed
df['Votes'] = pd.to_numeric(df['Votes'], errors='coerce')
```

```python
# Distribution of Aggregate Rating
plt.figure(figsize=(8, 5))
sns.histplot(df['Aggregate rating'], bins=20, kde=True)
plt.title("Distribution of Aggregate Rating")
plt.xlabel("Aggregate Rating")
plt.ylabel("Count")
plt.grid(True)
plt.show()
```

```python
# Class Imbalance
plt.figure(figsize=(10, 4))
sns.countplot(x='Aggregate rating', data=df, palette='viridis')
plt.title("Class Imbalance in Ratings")
plt.xlabel("Aggregate Rating")
plt.ylabel("Number of Restaurants")
plt.grid(axis='y')
plt.show()
```

task 2 :- Descriptive analysis


```python
print("\nðŸ”¹ Statistical Summary (Numerical Columns)")
print(df.describe())

print("\nðŸ”¹ Top 5 Country Codes:")
print(df['Country Code'].value_counts().head())

print("\nðŸ”¹ Top 5 Cities:")
print(df['City'].value_counts().head())

print("\nðŸ”¹ Top 5 Cuisines:")
print(df['Cuisines'].value_counts().head())
```

```python
# Top Cities Bar Plot
top_cities = df['City'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_cities.index, y=top_cities.values, palette="coolwarm")
plt.title("Top 10 Cities with Most Restaurants")
plt.ylabel("Number of Restaurants")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Top Cuisines Bar Plot
top_cuisines = df['Cuisines'].value_counts().head(10)
plt.figure(figsize=(10,5))
sns.barplot(x=top_cuisines.values, y=top_cuisines.index, palette="Set2")
plt.title("Top 10 Cuisines")
plt.xlabel("Number of Restaurants")
plt.grid(axis='x')
plt.show()
```

Task 3 :- Geospatial analysis


```python
# Plotting restaurant locations
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Longitude', y='Latitude', hue='Aggregate rating', data=df, palette='cool', alpha=0.7)
plt.title("Restaurant Locations by Rating")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.legend(title='Rating', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.show()
```

```python

# Count of restaurants per city
city_counts = df['City'].value_counts().head(10)
plt.figure(figsize=(10, 4))
sns.barplot(x=city_counts.index, y=city_counts.values, palette="mako")
plt.title("Top 10 Cities with Restaurants")
plt.ylabel("Restaurant Count")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```
